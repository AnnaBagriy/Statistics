```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
def load_dataset(filename):
    df = pd.read_excel(filename)

    return df
```

### Constants

```python
filename = 'kaffee.xls'
x_name = 'alter'
y_name = 'marke'

alpha = 0.05
```

### Data loading

```python
df = load_dataset(filename)

observed = pd.crosstab(df[x_name], df[y_name])
print(observed)
```

### Scatter plot

```python
rng = np.random.RandomState(0)

sizes = 1000 * rng.rand(df[y_name].size)

_ = plt.scatter(df[x_name], df[y_name], alpha=0.04, s=sizes)

_ = plt.xlabel(' ')
_ = plt.ylabel(' ')
```

### Chi-squared test

```python
statistics, p_value, dof, expected = st.chi2_contingency(observed)

critical = st.chi2.ppf(1 - alpha, dof)
```

```python
print('alpha =', alpha)
print('p_value =', p_value)
print()
print('statistics =', statistics)
print('critical =', critical)
```
